```python
!pip install vaderSentiment nltk matplotlib
import nltk
from nltk.tokenize import sent_tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Download necessary NLTK data
nltk.download('punkt')

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Example text data
text = """
Python is an amazing programming language. I love using it for data science.
Sometimes, debugging errors can be frustrating. However, the vast community support helps a lot.
"""

# Tokenize text into sentences
sentences = sent_tokenize(text)

# Analyze sentiment for each sentence
sentiments = []
for sentence in sentences:
    sentiment = analyzer.polarity_scores(sentence)
    sentiments.append(sentiment)
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {sentiment}\n")

# Extract individual sentiment scores for visualization
neg_scores = [s['neg'] for s in sentiments]
neu_scores = [s['neu'] for s in sentiments]
pos_scores = [s['pos'] for s in sentiments]
compound_scores = [s['compound'] for s in sentiments]

# Plot the sentiment scores
x = range(len(sentences))
plt.figure(figsize=(10, 6))
plt.plot(x, neg_scores, label='Negative', color='red')
plt.plot(x, neu_scores, label='Neutral', color='blue')
plt.plot(x, pos_scores, label='Positive', color='green')
plt.plot(x, compound_scores, label='Compound', color='purple', linestyle='--')
plt.xlabel('Sentence')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis of Each Sentence')
plt.legend()
plt.show()
```
